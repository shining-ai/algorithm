```python
import numpy as np
```

```python
N = 10
X = list(range(N + 1))
```

```python
X
```

```python
X_2 = [x ** 2 for x in X]
X_2
```

```python
log_X = [np.log(x) for x in X]
log_X
```

```python
X__2 = [2 ** x for x in X]
X__2
```

```python
from matplotlib import pyplot
import matplotlib.pyplot as plt
```

```python
plt.plot(X,X_2)
plt.plot(X,X)
plt.plot(X,log_X)
plt.plot(X,X__2)
plt.show()
```
